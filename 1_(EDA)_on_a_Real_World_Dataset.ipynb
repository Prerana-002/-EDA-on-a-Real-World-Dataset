```python
import pandas as pd

# Load the uploaded dataset
data = pd.read_csv('/content/DATASET AB_NYC_2019.csv')  # Ensure this filename matches the uploaded file

# Display the first few rows of the dataset
print(data.head())
```

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display the first few rows of the dataset
print(data.head())

# Data cleaning
data = data.dropna(subset=['price', 'neighbourhood_group'])
data = data.drop_duplicates()

# Distribution of Price
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=50, kde=True)
plt.title('Distribution of Airbnb Prices in NYC')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Price vs. Neighborhood Group
plt.figure(figsize=(10, 6))
sns.boxplot(x='neighbourhood_group', y='price', data=data)
plt.title('Price Distribution by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Price')
plt.show()
```

Project Title: Exploratory Data Analysis on Airbnb Listings in NYC

Objective: To analyze the Airbnb listings in NYC to uncover insights related to pricing, availability, and neighborhood characteristics.

Key Findings:

The average price of listings is skewed due to a small number of high-priced properties.
Most Airbnb listings are concentrated in neighborhoods like Manhattan and Brooklyn.
There is a positive correlation between the number of reviews and the price, indicating that more popular listings tend to be priced higher.
Visualizations:

Distribution of prices shows a right skew.
Box plots reveal significant outliers in pricing, particularly in certain neighborhoods.
A bar chart indicates that Brooklyn has the highest number of listings.
Conclusion: The analysis provides valuable insights for both hosts and potential guests, highlighting pricing trends and popular areas within NYC.
